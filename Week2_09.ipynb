2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16 dan int32?

  **Jawab:**
*   **int16:** Tipe data ini memiliki rentang nilai dari -32,768 hingga 32,767. Gambar hitam akan diwakili oleh nilai 0, tetapi ada risiko overflow jika operasi tertentu dilakukan.
*   **int32:** Tipe data ini memiliki rentang nilai yang lebih luas, yaitu dari -2,147,483,648 hingga 2,147,483,647. Ini memberikan lebih banyak fleksibilitas dalam pengolahan piksel.
*   **Pengaruh pada black image:** Ketika menggunakan tipe data int16 untuk menghasilkan gambar hitam, kualitas gambar mungkin sedikit berkurang, terutama pada saat perbesaran atau zoom, dan risiko terjadinya overflow atau underflow saat melakukan operasi matematika yang kompleks lebih tinggi. Sebaliknya, tipe data int32 menghasilkan gambar dengan kualitas yang lebih halus dan detail yang lebih kaya, serta memberikan ruang yang lebih besar untuk melakukan berbagai macam operasi matematika tanpa khawatir akan masalah overflow.

# **Tugas**
Tutup tiap bagian tertentu dari KTM Anda menggunakan fungsi yang telah Anda pelajari. Kreasikan terkait dengan warna dan ukuran dari bentuk-bentuknya.

*    Kelompok 9 tutup bagian Nama dan tanggal valid  


```python
import cv2 as cv
import matplotlib.pyplot as plt

# Daftar gambar
img_paths = ["/content/drive/MyDrive/SEMESTER 5/PCVK/Images/KTM Mila.jpeg",
             "/content/drive/MyDrive/SEMESTER 5/PCVK/Images/KTM Hana.jpeg",
             "/content/drive/MyDrive/SEMESTER 5/PCVK/Images/KTM Putra.jpeg"]

for img_path in img_paths:
    # Membaca gambar
    ktm = cv.imread(img_path)
    ktm = cv.resize(ktm, (480, 297))
    image_ktm_coret = ktm.copy()

    # Koordinat untuk area yang ingin ditutup
    coordinates_name = (102, 235, 300, 255)
    coordinates_valid = (248, 180, 330, 208)

    # Menutup bagian nama dengan lingkaran berjejer
    radius = 5  # Radius lingkaran
    thickness = -1  # Mengisi lingkaran dengan warna
    color_name = (0, 255, 0)  # Warna hijau dalam format BGR
    for x in range(coordinates_name[0], coordinates_name[2], radius * 2):
        for y in range(coordinates_name[1], coordinates_name[3], radius * 2):
            cv.circle(image_ktm_coret, (x, y), radius, color_name, thickness)

    # Menutup bagian tanggal valid dengan kotak
    color_valid = (0, 255, 255)  # Warna kuning dalam format BGR
    cv.rectangle(image_ktm_coret,
                 (coordinates_valid[0], coordinates_valid[1]),
                 (coordinates_valid[2], coordinates_valid[3]),
                 color_valid,  # Warna kotak
                 thickness=-1)  # Mengisi kotak dengan warna

    # Konversi gambar dari BGR ke RGB
    image_ktm_coret_rgb = cv.cvtColor(image_ktm_coret, cv.COLOR_BGR2RGB)
    ktm_rgb = cv.cvtColor(ktm, cv.COLOR_BGR2RGB)

    # Menampilkan gambar dengan Matplotlib dalam subplots
    plt.figure(figsize=(12, 6))

    # Menampilkan gambar asli
    plt.subplot(1, 2, 1)
    plt.imshow(ktm_rgb)
    plt.title('Gambar Asli')
    plt.axis('on')  # Menampilkan sumbu

    # Menampilkan gambar yang dimodifikasi
    plt.subplot(1, 2, 2)
    plt.imshow(image_ktm_coret_rgb)
    plt.title('Gambar Dimodifikasi')
    plt.axis('on')  # Menampilkan sumbu

    plt.show()
    print('\n')
```
